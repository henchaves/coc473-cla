```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_05 = pd.read_csv("runge_kutta_nystrom_0_5.csv", index_col=0)
df_05
```

```python
df_01 = pd.read_csv("runge_kutta_nystrom_0_1.csv", index_col=0)
df_01
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Deslocamento (m) x Tempo (s); h=0.5", fontsize=14)
ax.plot(df_05["tempo"], df_05["deslocamento"], color="tomato", label="h = 0.5")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Deslocamento (m)")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Deslocamento (m) x Tempo (s); h=0.1", fontsize=14)
ax.plot(df_01["tempo"], df_01["deslocamento"], color="steelblue", label="h = 0.1")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Deslocamento (m)")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Velocidade (m/s) x Tempo (s); h=0.5", fontsize=14)
ax.plot(df_05["tempo"], df_05["velocidade"], color="tomato", label="h = 0.5")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Velocidade (m/s)")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Velocidade (m/s) x Tempo (s); h=0.1", fontsize=14)
ax.plot(df_01["tempo"], df_01["velocidade"], color="steelblue", label="h = 0.1")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Velocidade (m/s)")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Aceleração (m/s²) x Tempo (s); h=0.5", fontsize=14)
ax.plot(df_05["tempo"], df_05["aceleração"], color="tomato", label="h = 0.5")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Aceleração (m/s²)")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.set_title("Aceleração (m/s²) x Tempo (s); h=0.1", fontsize=14)
ax.plot(df_01["tempo"], df_01["aceleração"], color="steelblue", label="h = 0.1")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Aceleração (m/s²)")
plt.legend()
plt.show()
```
